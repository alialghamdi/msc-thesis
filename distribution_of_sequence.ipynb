```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer

df = pd.read_csv('english_only.csv')

tokenizer = Tokenizer(num_words=None)
tokenizer.fit_on_texts(df['excerpt_value_cleaned'])
sequences = tokenizer.texts_to_sequences(df['excerpt_value_cleaned'])
sequence_lengths = [len(seq) for seq in sequences]


plt.figure(figsize=(10, 6))
plt.hist(sequence_lengths, bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Sequence Lengths')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.show()
percentile = 99
max_len = np.percentile(sequence_lengths, percentile)
print(f'Selected max_len at {percentile}th percentile: {int(max_len)}')
```

```python
import pandas as pd
import numpy as np
from transformers import BertTokenizer
import matplotlib.pyplot as plt


df = pd.read_csv('english_only.csv')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

tokenized_sequences = df['excerpt_value_cleaned'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True))
sequence_lengths = tokenized_sequences.apply(len)
percentile_99 = np.percentile(sequence_lengths, 99)
print(f'The 99th percentile of tokenized sequence lengths is: {int(percentile_99)}')

plt.figure(figsize=(10, 6))
plt.hist(sequence_lengths, bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Tokenized Sequence Lengths')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.axvline(percentile_99, color='red', linestyle='dashed', linewidth=1)
plt.text(percentile_99, plt.ylim()[1]*0.9, f'99th percentile: {int(percentile_99)}', color = 'red')
plt.show()
```

```python
import pandas as pd
import numpy as np
from transformers import RobertaTokenizer
import matplotlib.pyplot as plt


df = pd.read_csv('english_only.csv')  

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
tokenized_sequences = df['excerpt_value_cleaned'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True))
sequence_lengths = tokenized_sequences.apply(len)
percentile_99 = np.percentile(sequence_lengths, 99)
print(f'The 99th percentile of tokenized sequence lengths for RoBERTa is: {int(percentile_99)}')

plt.figure(figsize=(10, 6))
plt.hist(sequence_lengths, bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Tokenized Sequence Lengths for RoBERTa')
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.axvline(percentile_99, color='red', linestyle='dashed', linewidth=1)
plt.text(percentile_99, plt.ylim()[1]*0.9, f'99th percentile: {int(percentile_99)}', color = 'red')
plt.show()
```
